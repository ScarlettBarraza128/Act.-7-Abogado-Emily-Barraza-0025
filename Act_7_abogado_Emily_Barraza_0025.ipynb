```python
# =========================================
# SISTEMA DE IA PARA ABOGADOS Y JUZGADOS
# Nivel: Principiante
# =========================================

# Variables iniciales
sistema_activo = True
casos_resueltos = 0

print("======================================")
print("   IA DE ABOGADOS Y JUZGADOS")
print("======================================")

# Ciclo principal
while True:

    print("\n------ MENÚ PRINCIPAL ------")
    print("1. Analizar caso judicial")
    print("2. Ver casos resueltos")
    print("3. Salir")

    opcion = input("Seleccione una opción: ")

    # Opción 1
    if opcion == "1":

        print("\n--- NUEVO CASO JUDICIAL ---")

        acusado = input("Nombre del acusado: ")
        gravedad = int(input("Nivel de gravedad del delito (1-10): "))
        pruebas = input("¿Existen pruebas? (si/no): ")
        abogado = input("¿Tiene abogado defensor? (si/no): ")
        testigos = int(input("Cantidad de testigos en contra: "))

        print(f"\nAnalizando caso de {acusado}...")

        # Primera decisión
        if gravedad >= 8 and pruebas == "si":
            sentencia = 20
            print("Resultado: Delito muy grave.")

        elif gravedad >= 5:
            sentencia = 10
            print("Resultado: Delito moderado.")

        elif gravedad > 0:
            sentencia = 3
            print("Resultado: Delito leve.")

        else:
            sentencia = 0
            print("Resultado: Caso inválido.")

        # Defensa del abogado
        if abogado == "si":
            sentencia = sentencia - 2
            print("El abogado logró reducir la sentencia.")

        else:
            print("No hubo defensa legal.")

        # Revisión de testigos
        if testigos >= 3:
            sentencia = sentencia + 2
            print("Los testigos aumentaron la sentencia.")

        else:
            print("Pocos testigos en contra.")

        # Evitar números negativos
        if sentencia < 0:
            sentencia = 0

        print(f"\nSentencia final para {acusado}: {sentencia} años.")

        casos_resueltos = casos_resueltos + 1

    # Opción 2
    elif opcion == "2":

        print("\n--- REPORTE DEL SISTEMA ---")
        print("Casos resueltos:", casos_resueltos)

    # Opción 3
    elif opcion == "3":

        print("\nCerrando sistema judicial...")
        print("Gracias por usar la IA Judicial.")
        break

    # Opción inválida
    else:
        print("\nOpción no válida. Intente otra vez.")

print("Elaborado por Emily Barraza NL = 7 ABOGADO")
```
